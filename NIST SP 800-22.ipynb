```python
import math
# Frequency Test within a Block
def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def frequency_within_block_test(bit_sequence, block_size):
    # Calculate the number of blocks
    num_blocks = len(bit_sequence) // block_size
    
    # Initialize list to store frequencies
    frequencies = []
    
    # Iterate over blocks
    for i in range(num_blocks):
        # Extract block from the bit sequence
        block = bit_sequence[i * block_size : (i + 1) * block_size]
        
        # Calculate frequency of ones in the block
        num_ones = block.count('1')
        frequency = num_ones / block_size
        
        # Append frequency to list
        frequencies.append(frequency)
    
    # Calculate chi-square statistic
    chi_square = sum((freq - 0.5) ** 2 for freq in frequencies) * (4 * block_size)
    
    # Calculate p-value using the chi-square distribution
    p_value = math.erfc(math.sqrt(chi_square / 2))
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
block_size = 4  # Example block size
bit_sequence = bytes_to_bits(byte_sequence)
p_value = frequency_within_block_test(bit_sequence, block_size)
print("P-value:", p_value)
```

```python
# Tests for the Longest-Run-of-Ones in a Block
import math

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def longest_run_of_ones_in_block_test(bit_sequence, block_size):
    num_blocks = len(bit_sequence) // block_size
    max_run = 0
    
    for i in range(num_blocks):
        block = bit_sequence[i * block_size : (i + 1) * block_size]
        run = 0
        
        for bit in block:
            if bit == '1':
                run += 1
                max_run = max(max_run, run)
            else:
                run = 0
    
    # Calculate the expected maximum run for a block
    expected_max_run = (math.log(2 * block_size) / math.log(2)) + 2 * (math.log(block_size) / math.log(2)) - 2
    
    # Calculate the p-value using the normal approximation
    deviation = math.sqrt(2 * expected_max_run * (4 * block_size - 2) / (9 * block_size * (block_size - 1)))
    z_score = (max_run - expected_max_run) / deviation
    p_value = math.erfc(abs(z_score) / math.sqrt(2))
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
block_size = 8  # Block size in bits (1 byte)
p_value = longest_run_of_ones_in_block_test(bit_sequence, block_size)
print("P-value:", p_value)
```

```python
#The Binary Matrix Rank Test 
import numpy as np

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def binary_matrix_rank_test(bit_sequence, matrix_rows, matrix_cols):
    # Convert bit sequence to binary matrix
    num_bits = len(bit_sequence)
    num_matrices = num_bits // (matrix_rows * matrix_cols)
    binary_matrix = np.zeros((num_matrices, matrix_rows, matrix_cols), dtype=int)
    for i in range(num_matrices):
        for j in range(matrix_rows):
            for k in range(matrix_cols):
                binary_matrix[i, j, k] = int(bit_sequence[i * (matrix_rows * matrix_cols) + j * matrix_cols + k])
    
    # Compute ranks of submatrices
    ranks = []
    for matrix in binary_matrix:
        ranks.append(np.linalg.matrix_rank(matrix))
    
    # Compute proportion of full rank matrices
    num_full_rank = sum(rank == min(matrix_rows, matrix_cols) for rank in ranks)
    proportion_full_rank = num_full_rank / num_matrices
    
    return proportion_full_rank

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
matrix_rows = 8  # Number of rows in submatrices
matrix_cols = 8  # Number of columns in submatrices
proportion_full_rank = binary_matrix_rank_test(bit_sequence, matrix_rows, matrix_cols)
print("Proportion of full rank matrices:", proportion_full_rank)
```

```python
# The Discrete Fourier Transform (Spectral) Test
import numpy as np

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def discrete_fourier_transform_test(bit_sequence):
    n = len(bit_sequence)
    s = [int(bit) * 2 - 1 for bit in bit_sequence]  # Convert bits to +1 and -1
    
    # Compute the Discrete Fourier Transform (DFT)
    dft = np.fft.fft(s)
    
    # Compute the test statistic
    modulus_squared = np.abs(dft) ** 2
    T = (1 / n) * np.sum(modulus_squared)
    
    # Compute the p-value using the chi-square distribution
    p_value = np.exp(-T / 2)
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
p_value = discrete_fourier_transform_test(bit_sequence)
print("P-value:", p_value)
```

```python
# The Non-overlapping Template Matching Test
from scipy.stats import chi2

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def non_overlapping_template_matching_test(bit_sequence, template):
    # Count the occurrences of the template in the bit sequence
    num_blocks = len(bit_sequence) // len(template)
    count = 0
    for i in range(num_blocks):
        block = bit_sequence[i * len(template) : (i + 1) * len(template)]
        if block == template:
            count += 1
    
    # Compute the test statistic and the expected value
    n = len(bit_sequence)
    m = len(template)
    expected_count = (n - m + 1) / (2 ** m)
    chi_square = (count - expected_count) ** 2 / expected_count
    
    # Compute the p-value using the chi-square distribution
    p_value = 1 - chi2.cdf(chi_square, 1)
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
template = '1010'  # Example template
p_value = non_overlapping_template_matching_test(bit_sequence, template)
print("P-value:", p_value)
```

```python
# Maurer's "Universal Statistical" Test
from math import sqrt
from scipy.special import erfc
from collections import Counter

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def maurers_universal_test(bit_sequence, substring_length, block_length):
    # Calculate the number of substrings and blocks
    num_substrings = len(bit_sequence) // substring_length
    num_blocks = len(bit_sequence) // block_length
    
    # Initialize counters for substring frequencies
    substring_counts = Counter()
    
    # Count the occurrences of each substring
    for i in range(num_substrings):
        substring = bit_sequence[i * substring_length : (i + 1) * substring_length]
        substring_counts[substring] += 1
    
    # Calculate the test statistic
    sum_chi_square = 0
    for count in substring_counts.values():
        sum_chi_square += (count - num_blocks / 2) ** 2
    chi_square = (sum_chi_square - num_blocks / 2) / (num_blocks / 2 ** substring_length)
    
    # Calculate the p-value using the incomplete gamma function
    p_value = erfc(sqrt(chi_square / 2))
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
substring_length = 4  # Length of substrings to analyze
block_length = 16  # Length of blocks for counting substrings
p_value = maurers_universal_test(bit_sequence, substring_length, block_length)
print("P-value:", p_value)
```

```python
# The Approximate Entropy Test
from math import log
from collections import Counter

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def approximate_entropy_test(bit_sequence, pattern_length):
    n = len(bit_sequence)
    patterns = [bit_sequence[i:i+pattern_length] for i in range(n - pattern_length + 1)]
    
    # Count the occurrences of all possible patterns of the specified length
    pattern_counts = Counter(patterns)
    
    # Calculate probabilities
    probabilities = {pattern: count / (n - pattern_length + 1) for pattern, count in pattern_counts.items()}
    
    # Compute approximate entropy
    sum_entropy = sum(-prob * (pattern_length - 1) * (probabilities[pattern] - 1) for pattern, prob in probabilities.items())
    approximate_entropy = sum_entropy / (n - pattern_length + 1)
    
    # Compute the test statistic
    chi_square = 2 * n * (log(2) - approximate_entropy)
    
    # Compute the p-value using the chi-square distribution
    from scipy.stats import chi2
    p_value = 1 - chi2.cdf(chi_square, 2 ** (pattern_length - 1))
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
pattern_length = 8  # Length of the pattern to test
p_value = approximate_entropy_test(bit_sequence, pattern_length)
print("P-value", p_value)
```

```python
# The Cumulative Sums (Cusums) Test
def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def cusums_test(bit_sequence):
    n = len(bit_sequence)
    S = [0] * (n + 1)
    S[0] = 0

    # Calculate the binary sequence
    s = [int(bit) * 2 - 1 for bit in bit_sequence]

    for i in range(1, n + 1):
        S[i] = S[i - 1] + s[i - 1]

    max_S = max(S)
    min_S = min(S)

    # Compute the test statistic
    V_n = max(max_S, -min_S)

    # Compute the p-value using the asymptotic distribution
    from scipy.stats import norm
    p_value = 1 - norm.cdf(V_n * (1 - (4 * n + 10) / (9 * n * (2 * n - 1))))
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
p_value = cusums_test(bit_sequence)
print("P-value:", p_value)
```

```python
# The Random Excursions Test
from collections import Counter

def bytes_to_bits(byte_sequence):
    # Convert each byte to binary string and remove '0b' prefix
    binary_strings = [bin(byte)[2:].zfill(8) for byte in byte_sequence]
    
    # Concatenate all binary strings to form the bit sequence
    bit_sequence = ''.join(binary_strings)
    
    return bit_sequence

def random_excursions_test(bit_sequence):
    n = len(bit_sequence)
    x = [1 if bit == '1' else -1 for bit in bit_sequence]
    J = 6  # Number of states
    K = 8  # Length of excursions
    S = [0] * (n + 1)
    S[0] = 0

    # Calculate the cumulative sum
    for i in range(1, n + 1):
        S[i] = S[i - 1] + x[i - 1]

    cycles = [0] * J
    cycles[0] = 1
    state = 0

    for i in range(1, n + 1):
        state += x[i - 1]
        if state == 0:
            cycles[0] += 1
        if abs(state) < J:
            cycles[state] += 1

    # Compute the test statistic
    cycles[0] -= 1
    nu = [0] * J
    for i in range(J):
        if cycles[i] < K:
            nu[i] = 1
        else:
            nu[i] = cycles[i]

    # Compute the p-value using the Chi-Square distribution
    from scipy.stats import chi2
    chi_square = sum((nu[i] - cycles[i]) ** 2 / nu[i] for i in range(J))
    p_value = 1 - chi2.cdf(chi_square, J - 1)
    
    return p_value

# Example usage:
byte_sequence = b'\n\n\x9f*gr\x94%W\xabSU\xd7j\xf4B\xf8\xf6^\x01'  # Example byte sequence
bit_sequence = bytes_to_bits(byte_sequence)
p_value = random_excursions_test(bit_sequence)
print("P-value:", p_value)
```
